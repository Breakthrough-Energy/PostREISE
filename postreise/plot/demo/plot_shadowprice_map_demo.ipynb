```python
from postreise.plot.plot_shadowprice_map import plot_shadowprice
```

```python
# Plotting the shadowprice is easy! Just pick a scenario and a time
# Here we're plotting an hour with a large range of prices. 
# Note that plot_shadowprice() will choose split points automatically for you if you don't provide them.
# These are based on quantiles.
# However, if there are busses where the LMP is 0 or negative, we always split on those values

plot_shadowprice('91', '2016-08-04 23:00:00')
```

```python
# Here's a boring hour. The minimum LMP is greater than 0 so we don't bother splitting there.

plot_shadowprice('91', '2016-04-12 02:00:00')
```

```python
# You can even set your own split points. 
# Because we pick the colors by hand, the max number of split points is 10.
# Any busses outside the min and max value will not be shown

plot_shadowprice('91', '2016-04-12 02:00:00', [20,22,22.5,23,23.5,30])
```

```python
# We can also plot Texas and presumably other interconnects when we get there

plot_shadowprice('88', '2016-08-04 23:00:00')
```
