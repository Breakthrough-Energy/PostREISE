## Instantiate Representative Scenarios
The *2020 base scenario* along with the *Collaborative USA 2030 (extra transmission)* and *Eastern Offshore Wind, 2030 (extra transmission)* are considered here. The first one is presented in the scrolly telling at https://science.breakthroughenergy.org/ and the last two in the dashboard on the same website.

```python
scenario_1 = Scenario('824')
scenario_2 = Scenario('1149')
scenario_3 = Scenario('1338')
```

## Let's Run One at a Time

```python
plot_n_scenarios(scenario_1)
```

```python
plot_n_scenarios(scenario_2)
```

```python
plot_n_scenarios(scenario_3)
```

## Let's Plot Two at a Time

```python
plot_n_scenarios(scenario_1, scenario_2)
```

```python
plot_n_scenarios(scenario_1, scenario_3)
```

```python
plot_n_scenarios(scenario_2, scenario_3)
```

The code gracefully handles the situation when `'wind_offshore'` generators are meaningfully present (they exist and generate non-zero energy) in one scenario but not in another.

The scaling between carbon and energy is always calculated to be equal height for the left-most (first) scenario. This may result in odd plotting if the left-most scenario is not the most carbon-intensive one (tons carbon / MWh energy).

## Let's Plot Three at Once

```python
plot_n_scenarios(scenario_1, scenario_2, scenario_3)
```
