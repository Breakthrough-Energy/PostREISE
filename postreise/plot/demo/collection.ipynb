```python
from postreise.plot.analyze_set import fraction_renewable, fraction, generation
```

We consider scenarios ran for the California plan. The dictionary below must have the following structure:
* keys are labels that will be used for plotting;
* values are a `list` of scenario id. The list has two elements. The first one is the id of the scenario using the original grid (no increase in the capacity of the transmission lines) while the second one corresponds to the scenario using an enhanced grid (increase in the capacity of the transmission lines).  

```python
scenario = {'2016': ['0', None], '2020':['1', '2'], '2030': ['5', '6']}
```

## Generation Allotment for Solar Plants and Wind Farms

```python
fraction_renewable(scenario, 'California');
```

## Generation Allotment for California

```python
data_fraction = fraction(scenario, 'California')
print(data_fraction)
```

## Energy Produced for Various State

```python
zone = ['California', 'Washington', 'Montana', 'New Mexico']
data_generation = generation(scenario, zone, 0)
for df in data_generation:
    print(df.head(n=len(zone)+1))
```
