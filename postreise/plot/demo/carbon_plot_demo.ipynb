```python
from postreise.analyze import carbon
from powersimdata.scenario.scenario import Scenario
from postreise.plot import carbon_plot, carbon_plothelper, carbon_barchart
```

```python
s224 = Scenario('224') 
s87 = Scenario('87') 
```

```python
grid224 = s224.state.get_grid()
grid87 = s87.state.get_grid()
carbon224  =carbon.generate_carbon_stats(s224)
carbon87 = carbon.generate_carbon_stats(s87)
```

```python
carbon_224 = carbon.summarize_carbon_by_bus(carbon224, grid224.plant)
carbon_87 = carbon.summarize_carbon_by_bus(carbon87, grid87.plant)
```

```python
busmap224 = carbon_plothelper.make_busmap(grid224.bus, carbon_224)
busmap87 = carbon_plothelper.make_busmap(grid87.bus, carbon_87)
```

```python
carbon_plot.makemap_carbon(busmap224, 'AllMatchCA 2030')
```

```python
carbon_plot.makemap_carbon_compare(busmap87,  busmap224, "2016 base","2030 collaborative")
```

```python
carbon_barchart.carbon_barchart(carbon_87, carbon_224, '2016 base', '2030 all match CA')
```

```python
carbon_barchart.carbon_percentdiff(carbon_87, carbon_224)
```
