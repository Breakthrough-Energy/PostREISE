```python
from postreise.analyze import carbon
from powersimdata.scenario.scenario import Scenario
from postreise.plot import carbon_plot, carbon_plothelper, carbon_barchart
```

```python
s224 = Scenario('224') 
s87 = Scenario('87') 
```

```python
grid224 = s224.state.get_grid()
grid87 = s87.state.get_grid()
carbon224  = carbon.generate_carbon_stats(s224)
carbon87 = carbon.generate_carbon_stats(s87)
```

```python
carbon_224 = carbon.summarize_carbon_by_bus(carbon224, grid224.plant)
carbon_87 = carbon.summarize_carbon_by_bus(carbon87, grid87.plant)
```

```python
busmap224 = carbon_plothelper.make_emitter_map(grid224.bus, carbon_224)
busmap87 = carbon_plothelper.make_emitter_map(grid87.bus, carbon_87)
```

```python
carbon_plot.map_carbon_emission(busmap224, 'AllMatchCA 2030')
```

```python
_ = carbon_plot.map_carbon_emission_comparison(busmap87,  busmap224, "2016 base", "2030 collaborative")
```

```python
carbon_barchart.plot_carbon_bar(carbon_87, carbon_224, '2016 base', '2030 all match CA')
```

```python
carbon_barchart.print_carbon_diff(carbon_87, carbon_224)
```
