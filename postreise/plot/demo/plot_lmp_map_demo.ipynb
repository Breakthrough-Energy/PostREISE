```python
from powersimdata.scenario.scenario import Scenario  
from postreise.plot.plot_lmp_map import map_lmp
from bokeh.io import output_notebook, show
```

```python
#load scenario, scenario's grid, and lmps
s462 = Scenario('462')
s_grid = s462.state.get_grid()
lmp = s462.state.get_lmp()
```

```python
output_notebook()
show(map_lmp(s_grid,  lmp))
```
