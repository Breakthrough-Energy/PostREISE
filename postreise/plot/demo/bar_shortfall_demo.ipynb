```python
import matplotlib.pyplot as plt
from powersimdata.utility.helpers import PrintManager
from powersimdata.design.generation.clean_capacity_scaling import load_targets_from_csv

from postreise.plot.plot_bar_shortfall import plot_bar_shortfall
```

### Shortfall bar plot regular targets

```python
with PrintManager():
    target_df = load_targets_from_csv("../../data/2030_USA_Clean_Energy_Regular_Targets.csv")
    plot_bar_shortfall(
        ["Arizona", "Washington", "California", "New York", "all"], 
        [823, 824], 
        target_df,
        scenario_names=["USA 2016 Basecase", "USA 2020 Anchor"],
        baseline_scenario=823, 
        baseline_scenario_name="USA 2016 Basecase"
    )
plt.show()
```
