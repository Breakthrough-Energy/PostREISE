### Stacked Bar Plot

```python
scenario_ids = ['824', '1149', '1502', '1452', '1441', '1270']
scenario_names = [
    '2020',
    'Current\n2030 Goals',
    'Ambitious\n2030 Goals\nDesign 1',
    'Ambitious\n2030 Goals\nDesign 2a',
    'Ambitious\n2030 Goals\nDesign 2b',
    'Ambitious\n2030 Goals\nDesign 3'
]
```

```python
resources = [
    'curtailment', 
    'wind_curtailment', 
    'solar_curtailment', 
    'wind', 
    'solar', 
    'ng', 
    'coal', 
    'dfo', 
    'other', 
    'geothermal', 
    'hydro', 
    'nuclear'
]
```

```python
t2c = {
    'nuclear': '#173FA5',
    'hydro': '#0090FF',
    'geothermal': '#CC67F3',
    'other': '#8B36FF',
    'dfo': '#31E8CB',
    'coal': '#37404C',
    'ng': '#72818F',
    'solar': '#FFBB45',
    'wind': '#78D911',
    'curtailment': 'xkcd:red',
    'solar_curtailment': '#FFBB45',
    'wind_curtailment': '#78D911',
}

t2hc = {
    "solar_curtailment": "#996100",
    "wind_curtailment": "#4e8e0b",
}

t2l = {
    "nuclear": "Nuclear",
    "hydro": "Hydro",
    "geothermal": "Geothermal",
    "other": "Other",
    "dfo": "Distillate Fuel Oil",
    "coal": "Coal",
    "ng": "Natural Gas",
    "solar": "Solar",
    "wind": "Wind",
    "wind_offshore": "Wind Offshore",
    "biomass": "Biomass",
    "storage": "Storage",
    "wind_curtailment": "Wind Curtailment",
    "solar_curtailment": "Solar Curtailment"
}
```
