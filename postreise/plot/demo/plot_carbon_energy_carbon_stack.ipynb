#### Imports

```python
from postreise.plot.plot_energy_carbon_stack import plot_n_scenarios
from powersimdata.scenario.scenario import Scenario
```

#### Instantiate representative scenarios

```python
scenario_403 = Scenario('403')    #No 'wind_offshore' generators
scenario_511 = Scenario('511')    #Some 'wind_offshore' generators, but they're scaled to 0
scenario_571 = Scenario('571')    #Several 'wind_offshore' generators with non-zero capacities
```

#### Let's run one at a time

```python
plot_n_scenarios(scenario_403)
```

```python
plot_n_scenarios(scenario_511)
```

```python
plot_n_scenarios(scenario_571)
```

Even though `'wind_offshore'` generators are present in both 511 and 571, they are only plotted in 571 because for 511 they are all scaled down to 0 MW and therefore they generate no energy. There are no `'geothermal'` generators in any of these scenarios, because these generators are not present in Eastern.

#### Let's plot two at a time

```python
plot_n_scenarios(scenario_403, scenario_511)
```

```python
plot_n_scenarios(scenario_403, scenario_571)
```

```python
plot_n_scenarios(scenario_511, scenario_571)
```

The code gracefully handles the situation when `'wind_offshore'` generators are meaningfully present (they exist and generate non-zero energy) in one scenario but not in another.

The scaling between carbon and energy is always calculated to be equal height for the left-most (first) scenario. The may result in odd plotting if the left-most scenario is not the most carbon-intensive one (tons carbon / MWh energy).

#### Let's plot three at once

```python
plot_n_scenarios(scenario_403, scenario_511, scenario_571)
```
